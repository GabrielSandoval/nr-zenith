```python
# Data Crunching
import glob
import numpy as np
import pandas as pd
import re
import string

# Preprocessing, Feature Selection, Metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

# MODELS
from sklearn.neural_network import MLPClassifier

# Data Exportng
from joblib import dump, load

# Visualizations
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import seaborn as sns
%matplotlib inline

# Image Processing
from PIL import Image
```

```python
# Developer configurations
REPROCESS = 1
EXCLUDE_GLASSES = 1
```

```python
def show_confusion_matrix(cm):
    cmn = (cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]).round(4)
    fig,ax = plt.subplots(figsize=(3,3))
    ax.matshow(cmn,cmap=plt.cm.Oranges,alpha=0.85)
    for i in range(cmn.shape[0]):
        for j in range(cmn.shape[1]):
            ax.text(x=j,y=i,s=cmn[i,j],va='center',ha='center')
    plt.xlabel('predicted label')
    plt.ylabel('true label')
    plt.show()
```

```python
# Crawl over dataset in local file system
def crawl_images(path, outfile):
    paths = "{}/*/*.png".format(path)
    
    with open(outfile, "w") as file:
        headers = ["Image Path", "Subject ID", "Image ID", "Gender", "Glasses", "Eye State", "Reflections", "Lighting Conditions", "Sensor ID"]
        file.write(",".join(headers))
        file.write("\n")
            
        for filename in glob.glob(paths):
            chars = re.escape(string.punctuation)
            line = [filename] + re.sub(r'['+chars+']', ' ', filename).split()
            labels = [
                line[0],  # Image Path
                line[3],  # Subject ID
                line[5],  # Image ID
                line[6],  # Gender
                line[7],  # Glasses
                line[8],  # Eye State
                line[9],  # Reflections
                line[10], # Lighting Conditions
                line[11], # Sensor ID
            ]
            
            file.write(",".join(labels))
            file.write("\n")
            
    return outfile
```

```python
# Resize all training images
def resize_images(image_paths, processed_image_path):
    with open(processed_image_path, "w") as file:
        for image_path in data["Image Path"]:
            image = Image.open(image_path)
            image = image.resize(IMAGE_SIZE)
            image_arr = np.asarray(image).reshape(IMAGE_SIZE[0]*IMAGE_SIZE[1],)
            image_str = ",".join(map(str, image_arr))
            file.write(image_str)
            file.write("\n")
```

```python
# CONSTANTS
IMAGE_SIZE = (90, 90)
IMAGE_PATH = 'data/images'
LABEL_PATH = 'data/label.txt'
```

```python
# Sample Image

image = Image.open("data/images/{}".format('s0014/s0014_03817_0_0_1_0_0_01.png'))
image = image.resize(IMAGE_SIZE)
image = np.asarray(image)
imshow(image)
```

```python
if REPROCESS: crawl_images(IMAGE_PATH, LABEL_PATH)
```

```python
data = pd.read_csv(LABEL_PATH)
data.head()
```

```python
print("Total Number of Images: {}".format(len(data)))
```

```python
# Filter out images with eye glasses
if EXCLUDE_GLASSES: data = data[data["Glasses"] == 0]

print("Total Number of Images After Filtering: {}".format(len(data)))
```

```python
if REPROCESS:
    # Analyze Image resolution
    data["Image Height"] = data["Image Path"].apply(lambda x: np.asarray(Image.open(x)).shape[0])
    data["Image Width"] = data["Image Path"].apply(lambda x: np.asarray(Image.open(x)).shape[1])

    # Resolution Stats
    print("Image Height Stats (min, max, avg): {}".format((data["Image Height"].min(), data["Image Height"].max(), data["Image Height"].mean())))
    print("Image Width Stats (min, max, avg): {}".format((data["Image Width"].min(), data["Image Width"].max(), data["Image Width"].mean())))
```

```python
processed_image_path = "data/processed_images.csv"
if REPROCESS: data.to_csv('data/processed_data.csv', index=False)
if REPROCESS: resize_images(data["Image Path"], processed_image_path)
```

```python
images_df = pd.read_csv(processed_image_path, header=None)
images_df.head()
```

```python
OUTPUT = {
    "Model path": [],
    "Sample Size": [],
    "Training Size": [],
    "TP": [],
    "FP": [],
    "FN": [],
    "TN": [],
    "Accuracy": [],
    "F1-score": [],
}

X = np.array(images_df.values) / 255
Y = data["Eye State"]

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

model_name = "Neural Network"
estimator = MLPClassifier(hidden_layer_sizes=(128, 128), max_iter=5000, alpha= 1e-2, solver='adam', verbose=1)
estimator.fit(X_train,y_train)
predictions = estimator.predict(X_test)
f1score = f1_score(y_test, predictions).round(4)
accuracy = accuracy_score(y_test, predictions).round(4)
cm = confusion_matrix(y_test,predictions)
[[tn, fp], [fn, tp]] = cm

model_path = "models/{} - {}.joblib".format(model_name, f1score)
dump(estimator, model_path, protocol=3)

OUTPUT["Sample Size"].append(len(X))
OUTPUT["Training Size"].append(len(X_train))
OUTPUT["Model path"].append(model_path)
OUTPUT["TP"].append(tp)
OUTPUT["FP"].append(fp)
OUTPUT["FN"].append(fn)
OUTPUT["TN"].append(tn)
OUTPUT["Accuracy"].append(accuracy)
OUTPUT["F1-score"].append(f1score)
```

```python
# Model evaluation
show_confusion_matrix(cm)
output = pd.DataFrame(OUTPUT)
output.reset_index(drop=True)
```
